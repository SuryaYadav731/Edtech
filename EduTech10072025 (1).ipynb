### Role-Wise Application Distribution

```python
sns.set(style="whitegrid")
role_counts = df['Applied_Role'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=role_counts.index, y=role_counts.values, palette='Set2')

plt.title("Role-Wise Application Distribution", fontsize=16)
plt.xlabel("Applied Role", fontsize=12)
plt.ylabel("Number of Candidates", fontsize=12)
plt.savefig("Role-Wise Application Distribution")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

### Skill Frequency Among Freshers

```python
skill_counts = df['Skills'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=skill_counts.index, y=skill_counts.values, palette='Set3')
plt.title("Skill Frequency Among Freshers", fontsize=16)
plt.xlabel("Skills", fontsize=12)
plt.ylabel("Number of Candidates", fontsize=12)
plt.savefig("Skill Frequency Among Freshers.png")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

### Degree vs Specialization Distribution

```python
combo_counts = df.groupby(["Degree", "Specialization"]).size().reset_index(name="Count")
plt.figure(figsize=(12, 6))
sns.barplot(x="Degree", y="Count", hue="Specialization", data=combo_counts, palette="Set2")

plt.title("Degree vs Specialization Distribution", fontsize=16)
plt.xlabel("Degree", fontsize=12)
plt.ylabel("Number of Candidates", fontsize=12)
plt.legend(title="Specialization", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.savefig("Degree vs Specialization Distribution")
plt.tight_layout()
plt.show()
```

### Preferred Job Locations Among Freshers

```python
location_counts = df['Preferred_Location'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=location_counts.index, y=location_counts.values, palette='Pastel1')

plt.title("Preferred Job Locations Among Freshers", fontsize=14)
plt.xlabel("Location", fontsize=12)
plt.ylabel("Number of Candidates", fontsize=12)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig("preferred_locations.png", dpi=300)
plt.show()
```

### Application Status Summary

```python
status_counts = df['Status'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel2.colors)
plt.title("Application Status Summary", fontsize=14)

plt.savefig("application_status_summary.png", dpi=300)
plt.show()
```

### Skill vs Applied Role Distribution

```python
skill_role_counts = df.groupby(["Skills", "Applied_Role"]).size().reset_index(name="Count")
plt.figure(figsize=(12, 6))
sns.barplot(x="Skills", y="Count", hue="Applied_Role", data=skill_role_counts, palette="Set2")

plt.title("Skill vs Applied Role Distribution", fontsize=16)
plt.xlabel("Skills", fontsize=12)
plt.ylabel("Number of Applicants", fontsize=12)
plt.xticks(rotation=30)
plt.legend(title="Applied Role", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.savefig("skill_vs_applied_role.png", dpi=300)
plt.show()
```

### Degree vs Application Status

```python
degree_status = df.pivot_table(index='Degree', columns='Status', aggfunc='size', fill_value=0)
degree_status.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Pastel2')

plt.title("Degree vs Application Status")
plt.xlabel("Degree")
plt.ylabel("Number of Candidates")
plt.legend(title="Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("degree_vs_status.png", dpi=300)
plt.show()
```

```python
skill_status = df.pivot_table(index='Skills', columns='Status', aggfunc='size', fill_value=0)
skill_status.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set3')

plt.title("Skill vs Application Status")
plt.xlabel("Skills")
plt.ylabel("Number of Candidates")
plt.legend(title="Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("skill_vs_status.png", dpi=300)
plt.show()
```
